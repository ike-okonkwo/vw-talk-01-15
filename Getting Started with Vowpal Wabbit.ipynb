```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE, MDS
from sklearn import datasets
import pandas as pd
```

```python
# convert from pandas dataframe to vw format

def convert_to_vw_format(input, output, target_feature ='target'):
    '''
    Transfrom Pandas dataframe to Vowpal Wabbit format and write output to file
    Input Dataframe is expected to have the labels/target as the last column
    
    (pandas dataframe, string, string ) -> (vw_file_format)
    '''
    count = 0

    with open(output, 'w') as vw_file:
        for i in range(len(input)): 
            vw_file.write(str(input[target_feature].ix[i]) + " | ")
            for feature_name in input.columns[:-1]:
                count += 1         
                if isinstance(input[feature_name].ix[i], str):   # account for categorical features
                    vw_file.write(feature_name.replace(' ','_') + '_' + str(input[feature_name].ix[i]) + ':1')  
                else:
                    # convert spaces to underscore in feature names and write feature:value
                    vw_file.write(feature_name.replace(' ','_') + ':' + str(input[feature_name].ix[i]))  
                if  count < len(input.columns[:-1]): # get rid of the trailing white space after the last feature 
                    vw_file.write(" ")
                else:
                    pass
                if count%len(input.columns[:-1]) == 0:
                    vw_file.write("\n")
            count = 0
        
```

### IRIS

Let's Visualise IRIS data in 2D by using three different dimensionaloity reduction techniques

```python
iris = datasets.load_iris()
iris_tsne = TSNE(learning_rate=100).fit_transform(iris.data)
iris_pca = PCA().fit_transform(iris.data)
iris_mds = MDS().fit_transform(iris.data)
```

```python
figure(figsize=(10, 5))
subplot(131)
title('t-SNE')
scatter(iris_tsne[:, 0], iris_tsne[:, 1], c=iris.target)
subplot(132)
title('PCA')
scatter(iris_pca[:, 0], iris_pca[:, 1], c=iris.target)
subplot(133)
title('MDS')
scatter(iris_mds[:, 0], iris_mds[:, 1], c=iris.target)
```

```python
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_data.head()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
# split the data
iris_train, iris_test, iris_train_label, iris_test_label = train_test_split(iris_data, iris.target, test_size = 0.4 , random_state = 40)
```

```python
iris_data_train = pd.DataFrame(iris_train, columns=iris.feature_names)
iris_data_train['target'] = iris_train_label
iris_data_train.head()
```

```python
# Add 1 to the target column. For multi class classification problems , Vowpal Wabbit expects labels in the range [1....n]
iris_data_train['target'] = iris_data_train['target'] + 1 
iris_data_train.head()
```

```python
iris_data_test = pd.DataFrame(iris_test, columns=iris.feature_names)
iris_data_test['target'] = iris_test_label
iris_data_test.head()
```

```python
# Add 1 to the target column. For multi class classification problems , Vowpal Wabbit expects labels in the range [1....n]
iris_data_test['target'] = iris_data_test['target'] + 1 
iris_data_test.head()
```

```python
convert_to_vw_format(iris_data_train,'iris.train.vw', target_feature ='target')
convert_to_vw_format(iris_data_test,'iris.test.vw', target_feature ='target')
```

### MNIST

```python
digits = datasets.load_digits()
digits_tsne = TSNE(learning_rate=100).fit_transform(digits.data)
digits_pca = PCA(n_components=5).fit_transform(digits.data)
digits_mds = MDS(max_iter=100, n_init=1).fit_transform(digits.data)
```

Let's Visualise MNIST data in 2D by using three different dimensionaloity reduction techniques

```python
figure(figsize=(20, 10))
subplot(131)
title('t-SNE')
scatter(digits_tsne[:, 0], digits_tsne[:, 1], c=digits.target)
subplot(132)
title('PCA')
scatter(digits_pca[:, 0], digits_pca[:, 1], c=digits.target)
subplot(133)
title('MDS')
scatter(digits_mds[:, 0], digits_mds[:, 1], c=digits.target)
```
